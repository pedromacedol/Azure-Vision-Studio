# Azure - Face

### Introdução

O Face é um serviço de Detecção Facial, onde o algoritmo de IA analisa os traços faciais e mapea de acordo com os pontos de referência facial. 

Como destaque deixo a possibilidade de utilização em áreas que utilizem medidas faciais para realizar de procedimentos ou venda de produtos, como exemplo:

- Clínicas de Estética, Óticas, Clinicas Odontológicas e etc.

Para mais informações consulte a documentação e o guia de introdução:
- https://learn.microsoft.com/pt-br/azure/ai-services/computer-vision/overview-identity
- https://microsoftlearning.github.io/mslearn-ai-fundamentals/Instructions/Labs/04-face.html
