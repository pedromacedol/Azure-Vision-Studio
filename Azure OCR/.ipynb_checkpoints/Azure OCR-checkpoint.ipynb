# Azure - OCR

### Introdução

O Azure Optical Character Recognition(OCR) é um serviço de Marchine Learning que realiza a extração de textos em imagens e documentos, podendo ser utilizado para diversas áreas de atuações como:

- Leitura de rótulos, placas de rua, cartazes, livros, artigos e relatórios

Um ponto que destaco é a possibilidade de utilizar a ferramenta para automatizar desde processos pessoais, como extrair anotações de estudos, até a extração de dados de documentos em empresas, otimizando o tempo de execução de tarefas e garantindo economia financeira.

Para mais informações consulte a documentação e o guia de introdução:
- https://learn.microsoft.com/pt-br/azure/ai-services/computer-vision/overview-ocr
- https://microsoftlearning.github.io/mslearn-ai-fundamentals/Instructions/Labs/05-ocr.html
